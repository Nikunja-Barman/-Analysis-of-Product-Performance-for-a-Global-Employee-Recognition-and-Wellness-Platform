# Statistical Modelling

### Label_encoding

```python
# importing LabelEncoder class from the sklearn.preprocessing module 
# to encode categorical features as integer values. 

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
col = ['sender_id', 'sender_country', 'feed type', 
       'company_names', 'receiver_id', 'received_country','manager_id']
```

```python
# fitting the LabelEncoder and transforming the categorical features as integer values.

for c in col:
   df[c]=le.fit_transform(df[c])
```

### one-hot encoding

```python
# importing OneHotEncoder class from the sklearn.preprocessing module 
# to encode categorical features as binary vectors. 
# here, i have encoded all the dates to binary features

from sklearn.preprocessing import OneHotEncoder
df = pd.get_dummies(df,columns = ['send_date', 'receive_date'], drop_first = True)
```

```python
# encoded dataframe

df
```

#### Note:-  Assuming Points  Columns be the Target variable 

```python
# putting all the independent variables of dataframe in to X- array

X=df.drop(['points'], axis=1).values
X
```

```python
# putting the Target variable of dataframe into Y- array

Y=df['points'].values
Y
```

```python
# splitting the Independent and Target variable into Train-data and Test-date

from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(X,Y, test_size=0.2, random_state=1)
```

### Feature scaling

```python
# importing StandardScaler class from the sklearn.preprocessing module to standardize features.
# this is done before fitting a model because to ensure that all features are on a comparable scale.

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
# standardizing x_train data

x_train=sc.fit_transform(x_train)
```

```python
# standardizing x_test data

x_test=sc.transform(x_test)
```

### Logistic Regression

```python
# importing LogisticRegression class from the sklearn.linear_model module 
# to perform logistic regression for binary and multiclass classification problems.
# this model is a multi-class classification problem where the target variable is carrying multiple
# values, not the binary values. So, multi_class='multinomial' parameter has been used.

from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(multi_class='multinomial', solver='lbfgs')
```

```python
# fitting the model

classifier.fit(x_train, y_train)
```

```python
# predicting the y_pred values using x_test values to compare the y_pred vales with y_test values

y_pred=classifier.predict(x_test)
```

```python
y_pred
```

### Evaluation metrics

#### Evaluating the performance of the model

```python
# importing confusion_matrix function from the sklearn.metrics module 
# to evaluate the performance of a classification model by computing the confusion matrix. 
# confusion matrix summarizes the predictions made by a classification model against the actual test data.

from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_test, y_pred)
conf_matrix
```

#### As the target variable is Points column, so there are multiple values in the points column. Hence, it is a multi-class classification probelm leading to the generation of  so many values in the Confusion_matrix created in the previous cell

```python
# converting the Confusion_matrix generated above into a dataframe for better handling of so many
# results

conf_df = pd.DataFrame(conf_matrix, index=['True: 0', 'True: 1', 'True: 2', 'True: 3','True: 4',
'True: 5','True: 6','True: 7','True: 8','True: 9','True: 10','True: 11','True: 12','True: 13',
'True: 14','True: 15','True: 16','True: 17','True: 18','True: 19','True: 20','True: 21','True: 22'], 
                       columns=['Pred: 0', 'Pred: 1', 'Pred: 2', 'Pred: 3','Pred: 4',
'Pred: 5','Pred: 6','Pred: 7','Pred: 8','Pred: 9','Pred: 10','Pred: 11','Pred: 12','Pred: 13',
'Pred: 14','Pred: 15','Pred: 16','Pred: 17','Pred: 18','Pred: 19','Pred: 20','Pred: 21','Pred: 22'])
print(conf_df)
```

```python
# calculating number of instances in each class

num_instances = conf_matrix.sum(axis=1)
print('Number of instances:', num_instances)
```

```python
# calculating number of instances predicted for each class

num_pred_instances = conf_matrix.sum(axis=0)
print('Number of predicted instances:', num_pred_instances)
```

```python
# calculating the number of instances that were correctly classified for each class
# finding True Positive values

TP = np.diag(conf_matrix)
print('True positives:', TP)
```

```python
# Calculate the number of instances that were misclassified for each class
# finding False Positive, False Negative, True Negative

FP = conf_matrix.sum(axis=0) - TP
FN = conf_matrix.sum(axis=1) - TP
TN = conf_matrix.sum() - (TP + FP + FN)
print('False positives:', FP)
print('False negatives:', FN)
print('True negatives:', TN)
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
```

```python
# Binarize the labels
Y = label_binarize(Y, classes=[0, 1, 2])
n_classes = Y.shape[1]
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.5, random_state=0)
```

```python
# Fit the model using logistic regression
clf = OneVsRestClassifier(LogisticRegression())
y_score = clf.fit(X_train, y_train).predict_proba(X_test)
```

```python
# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
```

```python
# Compute micro-average ROC curve and ROC area
fpr["micro"], tpr["micro"], _ = roc_curve(y_test.ravel(), y_score.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])
```

```python
# Plot ROC curves for each class

plt.figure()
lw = 2
colors = ['red', 'green', 'blue']
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=lw,
             label='ROC curve of class {0} (area = {1:0.2f})'
             ''.format(i, roc_auc[i]))

# Plot micro-average ROC curve
plt.plot(fpr["micro"], tpr["micro"],
         label='micro-average ROC curve (area = {0:0.2f})'
         ''.format(roc_auc["micro"]),
         color='deeppink', linestyle=':', linewidth=4)

# Add labels and legend
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC curve for multi-class classification')
plt.legend(loc="lower right")
plt.show()
```
